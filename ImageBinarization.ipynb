# Image binarization

```python
from skimage.io import imread, imshow
from skimage import img_as_float
import numpy as np
%matplotlib inline
```

```python
i1 = imread('ultrasound.bmp', as_grey = True)
imshow(i1)
i1.dtype
```

```python
#i1 = img_as_float(i1)
for i in range(0, i1.shape[0]):
    for j in range(0, i1.shape[1]):
        if i1[i][j] > 0:
            i1[i][j] = 0
        else:
            i1[i][j] = 1
imshow(i1)
```

```python
i2 = imread('ultrasound.bmp', as_grey = True)
imshow(i2)
```

```python
for i in range(0, i2.shape[0]):
    for j in range(0, i2.shape[1]):
        if i2[i][j] > 0.3:
            i2[i][j] = 0
        else:
            i2[i][j] = 1
imshow(i2)
```

```python
i3 = imread('ultrasound_complex.bmp', as_grey = True)
imshow(i3)
```

```python
for i in range(0, i3.shape[0]):
    for j in range(0, i3.shape[1]):
        if i3[i][j] < 0.01:
            i3[i][j] = 0
        elif (i3[i][j] < 0.07) and (i3[i][j] > 0.01):
            i3[i][j] = 1
        elif i3[i][j] > 0.07:
            i3[i][j] = 0
imshow(i3)
```

```python
from skimage.morphology import dilation, erosion, opening, closing
```

```python
core = np.ones((5, 5))
i4 = dilation(i3, core)
imshow(i4)
```

```python
core = np.ones((5, 5))
i4 = erosion(i3, core)
imshow(i4)
```

```python
core = np.ones((7, 7))
i4 = opening(i3, core)
imshow(i4)
```

```python
core = np.ones((7, 7))
i4 = closing(i3, core)
imshow(i4)
```
